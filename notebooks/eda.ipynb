```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
from collections import Counter

import re
import emoji
import pandas as pd
from nltk.corpus import stopwords
from wordsegment import load, segment
import string

# Assuming your DataFrame is named 'df' and has columns 'text' and 'label'

def perform_eda(df):
    """
    Performs Exploratory Data Analysis (EDA) on a DataFrame containing text and labels.
    """

    print("--- DataFrame Information ---")
    print(df.info())

    print("\n--- Summary Statistics ---")
    print(df.describe())

    print("\n--- Label Distribution ---")
    print(df['label'].value_counts())

    plt.figure(figsize=(8, 6))
    sns.countplot(x='label', data=df)
    plt.title('Label Distribution')
    plt.show()

    # Text Length Analysis
    df['text_length'] = df['text'].apply(len)
    print("\n--- Text Length Statistics ---")
    print(df['text_length'].describe())

    plt.figure(figsize=(8, 6))
    sns.histplot(df['text_length'], bins=50, kde=True)
    plt.title('Text Length Distribution')
    plt.xlabel('Text Length')
    plt.show()

    # Word Count Analysis
    df['word_count'] = df['text'].apply(lambda x: len(str(x).split()))
    print("\n--- Word Count Statistics ---")
    print(df['word_count'].describe())

    plt.figure(figsize=(8, 6))
    sns.histplot(df['word_count'], bins=50, kde=True)
    plt.title('Word Count Distribution')
    plt.xlabel('Word Count')
    plt.show()

    # Simplified Language Detection (English vs. Tamil/Other)
    def detect_language_simple(text):
        english_count = 0
        tamil_count = 0
        for char in text:
            if ord(char) in range(65, 91) or ord(char) in range(97, 123):  # Basic English check
                english_count += 1
            elif ord(char) in range(2946, 3066): #Basic tamil check
                tamil_count +=1
        if english_count > tamil_count:
            return 'english/tanglish'
        else:
            return 'tamil'

    df['language'] = df['text'].apply(detect_language_simple)

    print("\n--- Language Distribution ---")
    print(df['language'].value_counts())

    plt.figure(figsize=(8, 6))
    sns.countplot(x='language', data=df)
    plt.title('Language Distribution')
    plt.show()

    # Common Words (Needs language aware stop words and tokenization)
    nltk.download('punkt', quiet=True) #For tokenization
    def get_common_words(texts, n=20):
        all_words = []
        for text in texts:
            words = word_tokenize(str(text).lower())
            all_words.extend(words)
        return Counter(all_words).most_common(n)

    def plot_common_words(word_counts, title="Common Words"):
        words, counts = zip(*word_counts)
        plt.figure(figsize=(12, 6))
        sns.barplot(x=list(counts), y=list(words))
        plt.title(title)
        plt.show()

    print("\n--- Common Words ---")
    common_words = get_common_words(df['text'])
    print(common_words)
    plot_common_words(common_words)

    # Common words by label
    for label in df['label'].unique():
        print(f"\n--- Common Words for Label: {label} ---")
        label_texts = df[df['label'] == label]['text']
        label_common_words = get_common_words(label_texts)
        print(label_common_words)
        plot_common_words(label_common_words, f"Common Words for Label: {label}")

    # Example of basic text cleaning (Needs significant improvement for mixed scripts)
    def clean_text(text):
        text = re.sub(r'[^a-zA-Z\u0B80-\u0C7F\s]', '', str(text)) #Keep letters and Tamil unicode range
        return text

    # -------------------------------
    # Text Preprocessing
    load()

    # Load stopwords
    tamil_stopwords = set(stopwords.words('tamil'))
    english_stopwords = set(stopwords.words('english'))

    # -------------------------------
    # Text Preprocessing
    # -------------------------------

    def preprocess_text(text):
        # Extract hashtags
        hashtags = re.findall(r"#\w+", text)

        # Process hashtags (convert to words)
        processed_hashtags = []
        for tag in hashtags:
            tag = tag.lstrip("#")
            words = segment(tag)
            processed_hashtags.append(" ".join(words))

        # Remove hashtags from main text
        text = re.sub(r"#\w+", "", text)

        # Replace mentions with a placeholder
        text = re.sub(r'@\w+', '<USER>', text)

        # Convert emojis to text
        text = emoji.demojize(text, delimiters=(" ", " "))
        text = text.translate(str.maketrans('', '', string.punctuation))
        # Remove extra whitespace
        text = re.sub(r'\s+', ' ', text).strip()

        # Merge cleaned text with processed hashtags
        final_text = text + " " + " ".join(processed_hashtags)

        return final_text

    # Function to remove stopwords
    def remove_stopwords(text):
        words = text.split()
        filtered = [word for word in words if word.lower() not in english_stopwords and word.lower() not in tamil_stopwords]
        return ' '.join(filtered)

    # Apply preprocessing
    df['cleaned_text'] = df['text'].apply(preprocess_text).apply(remove_stopwords)

    print("\n--- Example Cleaned Text ---")
    print(df[['text', 'cleaned_text']].head())

    print("\n--- Common Words After Cleaning ---")
    cleaned_common_words = get_common_words(df['cleaned_text'])
    print(cleaned_common_words)
    plot_common_words(cleaned_common_words, "Common Words After Cleaning")

    # Common words by label
    for label in df['label'].unique():
        print(f"\n--- Common Words for Label: {label} ---")
        label_texts = df[df['label'] == label]['cleaned_text']
        label_cleaned_common_words = get_common_words(label_texts)
        print(label_cleaned_common_words)
        plot_common_words(label_cleaned_common_words, f"Common Words for Label: {label} (Cleaned)")

perform_eda(df)
```

```python
import re

def categorize_language(text):
    # Tamil script (Unicode range: 0x0B80-0x0BFF)
    if re.search(r'[\u0B80-\u0BFF]', text):
        return 'Tamil'
    # English script (Unicode range: 0x0041-0x007A, 0x0061-0x007A)
    elif re.search(r'[a-zA-Z]', text):
            return 'Latin English'

df['language'] = df['text'].apply(categorize_language)
df['language'].value_counts()  # Get the distribution of languages
```
